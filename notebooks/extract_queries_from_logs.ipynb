# Extract queries from Virtuoso logs 

```python
import json
import os
import re
import urllib.parse


def parse_log_file(file_path):
    """Parse a single log file"""
    with open(file_path) as file:
        log_entries = file.readlines()
    queries_dict = {}

    log_pattern = re.compile(
        r'(?P<ip>\d+\.\d+\.\d+\.\d+) - - \[(?P<date>.*?)\] "GET /sparql\?query=(?P<query>.*?) HTTP.*?" (?P<status>\d+) \d+ ".*?" "(?P<user_agent>.*?)"'
    )

    for entry in log_entries:
        match = log_pattern.match(entry)
        if match:
            query_encoded = match.group("query").split("&")[0]  # Get the query part only
            query_decoded = urllib.parse.unquote_plus(query_encoded)
            user_agent = match.group("user_agent")
            queries_dict[query_decoded.strip()] = user_agent

    return queries_dict


def parse_log_directory(directory_path):
    queries_dict = {}
    for filename in os.listdir(directory_path):
        if filename.endswith(".log"):
            file_path = os.path.join(directory_path, filename)
            queries_dict = {**queries_dict, **parse_log_file(file_path)}
    return queries_dict


def fix_oneline_query(query: str) -> str:
    """Fix a query to be on multiple lines based on keywords"""
    if "\n" in query:
        return query
    newline_keywords = [
        "PREFIX",
        "SELECT",
        "WHERE",
        "where",
        "graph",
        "UNION",
        "FILTER",
        "OPTIONAL",
        "LIMIT",
        "ORDER BY",
        "GROUP BY",
    ]
    pattern = re.compile(r"\b(" + "|".join(newline_keywords) + r")\b")
    parts = pattern.split(query)
    new_lines = []
    for part in parts:
        if part.strip():
            if part.strip() in newline_keywords:
                new_lines.append(part.strip())
            else:
                if new_lines and new_lines[-1].strip() in newline_keywords:
                    new_lines[-1] += " " + part.strip()
                else:
                    new_lines.append(part.strip())
    return "\n".join(new_lines)


log_directory = "../data/logs-virtuoso"
# Order from longest query to shortest, could be improved with a more complex sorting based on BGP count
# But it is enough for now
queries_dict = dict(sorted(parse_log_directory(log_directory).items(), key=lambda item: -len(item[0])))

agent_count = {}
count = 0
for query, agent in queries_dict.items():
    if (
        not agent.startswith("Java-http-client")
        and "nonsensical.com" not in query
        and "graph <http://www.w3.org/ns/sparql-service-description>" not in query
        and "graph <http://www.openlinksw.com/schemas/virtrdf#>" not in query
    ):
        # Do not print queries from java void-generator
        print(f"{fix_oneline_query(query)}\n[{agent}]\n")
        count += 1
    if agent not in agent_count:
        agent_count[agent] = 1
    else:
        agent_count[agent] += 1

print(f"{count} queries found in logs")

agent_count = dict(sorted(agent_count.items(), key=lambda item: -item[1]))
print(json.dumps(agent_count, indent=2))
```
